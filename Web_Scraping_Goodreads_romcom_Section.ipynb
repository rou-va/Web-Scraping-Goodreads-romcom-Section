## Geting all Libraries ready

```python
# Importing Important Libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#load the webpage content
url= "https://www.goodreads.com/shelf/show/romcom"
page= requests.get(url)

# check status code for response received
# success code- 200
print (page)

# print content of request
print(page.content)
```

```python
#convert to beautiful soup
soup= bs(page.content, 'html.parser')

#print our web page
print(soup.prettify())
```

```python
# Extract all text from a page
print(soup.get_text())
```

## Web Scraping Begins...

```python
titles= soup.find_all('a', class_= 'bookTitle')
authors= soup.find_all('a', class_= 'authorName')
print(titles)
```

```python
# 1st method for removing tags(Title)
for title in titles:
    main_titles= title.get_text()
    print(main_titles)
```

```python
#2nd method for removing tags(Authors)
for author in authors:
    print(author.text)
```

```python
# Printing the titles and authors

for title, author in zip(titles, authors):
    print("Book: ", title.get_text(), "By: ", author.get_text())

    
```

```python
# Print title and authors 2
for title, author in zip(titles, authors):
    booklist= print("Book: ", title.get_text(),"By: ", author.get_text())
    
```

```python
# Extarcting links
# Find all anchor tags with "href"

for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# Extracting Images
images_list= []
images= soup.select('img')
for image in images:
    alt= image.get('alt')
    src= image.get('src')
    images_list.append({"alt":alt, "src": src})
    
for image in images_list:
    print(image)
```
